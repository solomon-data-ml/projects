## Find Area Under the Cos curve from -5 to 5

Approach:
1. Define the function
2. Define upper, lower limit and Delta 
3. calculate the f(x) in our case "cos" function from the lower value to upper value 
   by increasing to amout of delta

```python

import numpy as np
import matplotlib.pyplot as plot

 
x = np.arange(-10, 10, 0.5);
y = np.cos(x)

plot.plot(x,y);


```

```python
import math
```

```python
def f(x):
    return abs(math.cos(x))
     
```

```python
#Test the function
f(-4.8)
```

```python
lowerLimit = -5
upperLimit = 5

delta = 0.1
area = 0

while(lowerLimit<=upperLimit):
    area = area + (f(lowerLimit)*delta)
    lowerLimit = lowerLimit + delta
    
print(area)
```

To get more accurate result, choose the delta value as small as possible, but 
we should also consider the run time complexity.
We are calculating the area under the curve, and not doing integration and hence
we take the absolute value of the function.
